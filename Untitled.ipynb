```python
stock_dim = len(ticker_list)
state_space = 1 + 2*stock_dim + len(technical_indicators)*stock_dim
print(f"Stock Dimension: {stock_dim}, State Space: {state_space}")
```

```python
env_kwargs = {
    "hmax": 5000, 
    "initial_amount": 100000, 
    "buy_cost_pct": 0.001,
    "sell_cost_pct": 0.001,
    #"transaction_fee_percent": 1e-3,
    "state_space": state_space, 
    "stock_dim": stock_dim, 
    "tech_indicator_list": technical_indicators, 
    "min_buy_amount" : [1, 0, 5, 4],
    #"stock_attributes": technical_indicators,
    #"initial": True
    "action_space": stock_dim, 
    "reward_scaling": 1,
    "min_buy_amount": [1, 0, 5, 4],
    "min_transaction_amount": 50,
    "stoploss_penalty" : 0.7,
    "profit_loss_ratio" : 2,
    "cash_penalty_proportion" : 0.1,
}

e_train_gym = CryptoTradingEnv(df= train, **env_kwargs)
print(e_train_gym)
```

```python
env_train, _ = e_train_gym.get_sb_env()
print(type(env_train))
```

```python
e_trade_gym = CryptoTradingEnv(df = trade, **env_kwargs)
```

```python
net_archs = [[64,64], [96, 96], [128, 128], [64, 128, 128], [96, 128, 128]]
```

```python
def get_trained_model(params, model_type):
  e_train_gym = CryptoTradingEnv(df= train, **env_kwargs)
  env_train, _ = e_train_gym.get_sb_env()
  agent = DRLAgent(env = env_train)
  params['batch_size'] = int(params['batch_size'])
  params['n_steps'] = params['batch_size'] * 30 
  if 'net' in params: 
    net_params = dict(activation_fn=th.nn.Tanh, net_arch=net_archs[params['net']])
    params.pop('net')
    model_ = agent.get_model(model_type, model_kwargs= params, policy_kwargs=net_params)
    trained_model = agent.train_model(model= model_, tb_log_name= model_type, total_timesteps= 50000)
  else:
    model_ = agent.get_model(model_type, model_kwargs= params)
    trained_model = agent.train_model(model= model_, tb_log_name= model_type, total_timesteps= 50000)
  
  return trained_model
```

```python
best_ppo = {'batch_size': 72,
 'clip_range': 0.2,
 'ent_coef': 0.03,
 'gae_lambda': 0.97,
 'gamma': 0.99,
 'learning_rate': 0.03,
 'net': 0,
 'vf_coef': 0.25
           }
```

```python
best_model = get_trained_model(best_ppo, 'ppo')
ret, act, trades, cost = DRLAgent.DRL_prediction(model= best_model, environment= e_trade_gym)
```
